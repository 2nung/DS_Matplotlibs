```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
c_arr = ['r', 'g', 'b']
s_arr = [500, 2000, 5000]

x_loc = [-2, 0, 2]
y_loc = x_loc

fig, ax = plt.subplots(figsize=(10, 10))
for plot_idx in range(len(c_arr)):
    ax.scatter(x_loc[plot_idx], y_loc[plot_idx],
                c=c_arr[plot_idx],
                s=s_arr[plot_idx],
                label=' data ' + str(plot_idx))
ax.legend(loc='center left',
            bbox_to_anchor=(1.05, 0.5),
            labelspacing=10,
            fontsize=20,
            edgecolor='None')
```

```python
c_arr = ['r', 'g', 'b']
s_arr = [500, 2000, 5000]

x_loc = [-2, 0, 2]
y_loc = x_loc
fig, ax = plt.subplots(figsize=(10, 10))
for plot_idx in range(len(c_arr)):
    ax.scatter([], [],
                c=c_arr[plot_idx],
                s=s_arr[plot_idx],
                label=' data ' + str(plot_idx))

ax.legend(loc='center left',
            bbox_to_anchor=(1.05, 0.5),
            labelspacing=10,
            fontsize=20,
            edgecolor='None')
```

```python
legend_sizes = np.array([10, 20, 30, 40, 50])
fig, ax = plt.subplots(figsize=(10, 10))
for legend_size in legend_sizes:
    ax.scatter([], [],
                s=legend_size**2,
                c='tab:blue',
                label=str(legend_size))

ax.legend(loc='center left',
            bbox_to_anchor=(1, 0.5),
            title='Point Size',
            title_fontsize=30,
            fontsize=30,
            labelspacing=1)
```

```python
legend_sizes = np.array([10, 20, 30, 40, 50])

fig, ax = plt.subplots(figsize=(10, 10))
for legend_size in legend_sizes:
    ax.scatter([], [],
                facecolor='None',
                edgecolor='blue',
                linewidth=3,
                s=legend_size**2,
                label=str(legend_size))

ax.legend(loc='center left',
            bbox_to_anchor=(1, 0.5),
            title='Point Size',
            title_fontsize=30,
            fontsize=30,
            labelspacing=1)
```

```python
n_data = 100
x_data = np.random.normal(0, 3, (n_data,))
y_data = np.random.normal(0, 3, (n_data,))
s_arr = np.random.uniform(10, 50, (n_data,))

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x_data, y_data,
            s=s_arr**2,
            facecolor='None',
            edgecolor='blue',
            linewidth=4)
legend_sizes = np.array([10, 20, 30, 40, 50])
for legend_size in legend_sizes:
    ax.scatter([], [],
                facecolor='None',
                edgecolor='blue',
                linewidth=3,
                s=legend_size**2,
                label=str(legend_size))
ax.legend(loc='center left',
            bbox_to_anchor=(1, 0.5),
            title='Point Size',
            title_fontsize=30,
            fontsize=30,
            labelspacing=1)
```

```python
np.random.seed(0)
n_class = 5
n_data = 100
cmap = cm.get_cmap('tab20')
colors = [cmap(i) for i in range(n_class)]

fig, ax = plt.subplots(figsize=(10, 10))
center_pt = np.random.uniform(-10, 10, (n_class, 2))
for class_idx in range(n_class):
    center = center_pt[class_idx]
    x_data = center[0] + np.random.normal(0, 1, (n_data,))
    y_data = center[1] + np.random.normal(0, 1, (n_data,))
    s_arr = np.random.uniform(10, 40, (n_data,))
    ax.scatter(x_data, y_data,
                s=s_arr**2,
                facecolor='None',
                edgecolor=colors[class_idx],
                label='class ' + str(class_idx))
ax.legend(loc='upper right',
            fontsize=25)
```

```python
ax2 = ax.twinx()
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)

legend_sizes = [10, 20, 30, 40]
for size_idx, legend_size in enumerate(legend_sizes):
    ax2.scatter([], [],
        facecolor='None',
        edgecolor='black',
        linewidth=3,
        s=legend_size**2,
        label=str(legend_size))
ax2.legend(loc='center left',
            bbox_to_anchor=(1, 0.5),
            title='Point Size',
            title_fontsize=30,
            fontsize=30,
            labelspacing=1)
```
