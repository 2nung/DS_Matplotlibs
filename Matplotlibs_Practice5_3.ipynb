```python
np.random.seed(0)
n_data1, n_data2 = 300, 300

data1 = np.random.normal(0, 1, (n_data1,1))
data2 = np.random.normal(5, 3, (n_data2,1))

fig, ax = plt.subplots(figsize=(14, 10))
ax.tick_params(labelsize=20)

data = np.hstack((data1, data2))
m_val, M_val = data.min(), data.max()

data_loc = np.linspace(m_val, M_val, 31)
data_center = (data_loc[:-1] + data_loc[1:])/2
xticks = data_center[::2]

ax.hist(data1, bins=data_loc,
alpha=0.5)
ax.hist(data2, bins=data_loc,
alpha=0.5)

ax.set_xticks(xticks)
ax.set_xticklabels(np.around(xticks, 2),
rotation=30)

ax.set_xticks(data_center,
minor=True)

ax.grid()
ax.grid(which='minor',linestyle=':')
```

```python
np.random.seed(0)
n_class = 2
n_data = 300
data = np.random.normal(0, 1, (n_data, n_class))
fig, ax = plt.subplots(figsize=(14, 10))
ax.tick_params(labelsize=20)

freqs, bin_edges, rects = ax.hist(data)

bin_center = (bin_edges[:-1] + bin_edges[1:])/2
ax.set_xticks(bin_center)
ax.set_xticklabels(np.around(bin_center, 2),
                                rotation=30)
ylim = ax.get_ylim()
ax.set_ylim([ylim[0], ylim[1]+10])
```

```python
for rect in rects[0]:
    x = rect.get_x()
    width = rect.get_width()
    height = rect.get_height()
    ax.text(x + width/2, height+3,
            str(height),
            rotation=90,
            fontsize=20,
            ha='center',
            color='b')

for rect in rects[1]:
    x = rect.get_x()
    width = rect.get_width()
    height = rect.get_height()
    ax.text(x + width/2, height+3,
            str(height),
            rotation=90,
            fontsize=20,
            ha='center',
            color='r')
```

```python
colors = ['b', 'r']
for class_idx in range(n_class):
    for rect in rects[class_idx]:
        x = rect.get_x()
        width = rect.get_width()
        height = rect.get_height()
        ax.text(x + width/2, height+3,
                str(height),
                rotation=90,
                fontsize=20,
                ha='center',
                color=colors[class_idx])
```
