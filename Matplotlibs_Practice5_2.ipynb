```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
np.random.seed(0)
n_data = 500
data = np.random.normal(0, 1, (n_data,))
fig, ax = plt.subplots(figsize=(14, 10))
ax.tick_params(labelsize=30,
                length=10,
                width=3)
ax.hist(data,
        bins=20,
        facecolor='coral',
        edgecolor='red',
        linewidth=3)
ax.hist(data,
        facecolor='skyblue',
        edgecolor='darkblue',
        linewidth=3)
```

```python
np.random.seed(0)
n_data = 500
data = np.random.normal(0, 1, (n_data,))
fig, ax = plt.subplots(figsize=(14, 10))
ax.tick_params(labelsize=20)
freqs, bin_edges, rects = ax.hist(data,density=True)
bin_interval = bin_edges[1] - bin_edges[0]
max_freq = freqs.max()
ax.set_ylim([0, max_freq*bin_interval+0.05])
for rect in rects:
    height = rect.get_height()
    height *= bin_interval
    rect.set_height(height)
```

```python
np.random.seed(0)
n_data = 500
data = np.random.normal(0, 1, (n_data,))
fig, ax = plt.subplots(figsize=(14, 10))
ax.tick_params(labelsize=20)
freqs, bin_edges, rects = ax.hist(data)
bin_centers = (bin_edges[:-1] + bin_edges[1:])/2
ax.set_xticks(bin_centers)
ax.set_xticks(bin_edges,minor=True)
ax.grid(axis='x', which='major',linewidth=3)
ax.grid(axis='x', which='minor',linewidth=3,linestyle=':')
```
