```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
```

```python
PI = np.pi
t = np.linspace(0, 4*PI, 40)
cos = np.cos(t)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(10, 5))

ax.tick_params(labelsize=20)
ax.stem(cos,use_line_collection=True)
#ax.stem(t, sin,use_line_collection=True)
#ax.stem(t, cos,use_line_collection=True)
```

```python
ax.stem(t, cos,use_line_collection=True,linefmt='r')
ax.stem(t, cos,use_line_collection=True,linefmt='b')
ax.stem(t, cos,use_line_collection=True,linefmt='magenta')
ax.stem(t, sin,use_line_collection=True,linefmt='darkblue')
ax.stem(t, sin,use_line_collection=True,linefmt='C0')
ax.stem(t, sin,use_line_collection=True,linefmt='C1')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.tick_params(labelsize=20)

ax.stem(t, cos,use_line_collection=True,linefmt='C0:')
ax.stem(t, sin,use_line_collection=True,linefmt='C1-.')

```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.tick_params(labelsize=20)
ax.stem(t, cos,use_line_collection=True,linefmt='C0',markerfmt='C0o')
ax.stem(t, sin,use_line_collection=True,linefmt='C1',markerfmt='C1x')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.tick_params(labelsize=20)

ax.stem(t, cos,
        use_line_collection=True,
        linefmt='C0',
        markerfmt='C0o',
        basefmt='k')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.tick_params(labelsize=20)

ax.stem(t, cos,
        use_line_collection=True,
        linefmt='C0',
        markerfmt='C0o',
        basefmt='k',bottom=0.5)

ax.stem(t, sin,
        use_line_collection=True,
        linefmt='C1',
        markerfmt='C1x',
        basefmt='k',
        bottom=-0.5)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.tick_params(labelsize=20)

ax.stem(cos,
        linefmt='C1-',
        markerfmt='C1o',
        basefmt='k',
        use_line_collection=True,
        label='cos')
ax.stem(sin,
        linefmt='C2-',
        markerfmt='C2o',
        basefmt='k',
        use_line_collection=True,
        label='sin')
ax.legend(loc='upper left',
            bbox_to_anchor=(1, 1),
            fontsize=30)
fig.tight_layout()
```

```python
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position(('data', 0))

major_yticks = np.arange(-1, 1.01, 0.5)
minor_yticks = np.arange(-1, 1.01, 0.25)
ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks,minor=True)

ax.grid(axis='y')
ax.grid(axis='y',
        which='minor',
        linewidth=2,
        linestyle=':')
ax.legend(loc='upper center',
            bbox_to_anchor=(0.5, 0),
            ncol=2,
            fontsize=30)
fig.tight_layout()
```
