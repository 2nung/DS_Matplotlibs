```python



n_can = 5
votes = np.random.randint(100, 1000, (n_can, ))
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_vote = np.sum(votes)
percentages = votes/total_vote*100
       
sort_idx = np.argsort(percentages)
percentages_sort = percentages[sort_idx]
cans_sort = cans[sort_idx]

fig, ax = plt.subplots(figsize=(10, 10),
                        facecolor='lightsteelblue')

textprops = {'fontsize': 30, 'color': 'w'}
ax.pie(percentages_sort,
        autopct='%1.1f%%',
        textprops=textprops)

#ax.pie(percentages_sort,autopct='%1.1f%%',textprops=textprops,startangle=90)
```

```python
wedgeprops = {'edgecolor': 'lightsteelblue', 'linewidth':2}
cmap = cm.get_cmap('tab10')
ax.pie(percentages_sort,
        textprops=textprops,
        wedgeprops=wedgeprops,
        startangle=90,
        autopct='%1.1f%%')
colors = [cmap(i) for i in range(5)]
fig, ax = plt.subplots(figsize=(10, 10),
                        facecolor='lightsteelblue')
ax.pie(percentages_sort,
        textprops=textprops,
        wedgeprops=wedgeprops,
        startangle=90,
        autopct='%1.1f%%',
        colors=colors)
```

```python
fig, ax = plt.subplots(figsize=(10, 10),
facecolor='lightsteelblue')

explode = [0, 0, 0, 0, 0.1]

ax.pie(percentages_sort,
        colors=colors,
        explode=explode,
        textprops=textprops,
        wedgeprops=wedgeprops,
        startangle=90,
        autopct='%1.1f%%')
```

```python
wedges, texts, autotexts = ax.pie(percentages_sort,
                                    colors=colors,
                                    explode=explode,
                                    textprops=textprops,
                                    wedgeprops=wedgeprops,
                                    startangle=90,
                                    autopct='%1.1f%%')
ax.legend(wedges[::-1], cans_sort[::-1],
            title='Candidate',
            title_fontsize=20,
            fontsize=20,
            loc='center left',
            bbox_to_anchor=(0.9, 0.5))
fig.tight_layout()
```

```python
n_can = 5
votes = np.random.randint(100, 1000, (2, n_can))
votes_sum = np.sum(votes, axis=0)
cans = np.array(['A', 'B', 'C', 'D', 'E'])
total_vote = np.sum(votes_sum)
percentages = votes_sum/total_vote*100
percentages_tmp = votes/total_vote*100
```

```python
fig, ax = plt.subplots(figsize=(10, 10),
                        facecolor='lightsteelblue')

cmap = cm.get_cmap(name='tab20c', lut=20)
outer_colors = [cmap(4*i) for i in range(5)]
textprops = {'fontsize': 30, 'color': 'w'}
wedgeprops = {'width': 0.4,
'edgecolor': 'lightsteelblue',
'linewidth':2}

wedges, texts, autotexts = ax.pie(percentages,
colors=outer_colors,
textprops=textprops,
wedgeprops=wedgeprops,
pctdistance=0.8,
autopct='%1.1f%%')

```

```python
inner_colors = [cmap(4*i+j+1) for i in range(5)
                for j in range(2)]
wedgeprops = {'width': 0.3, 'edgecolor': 'lightsteelblue', 'linewidth':2}
ax.pie(percentages_tmp.flatten(order='F'),
        colors=inner_colors,
        wedgeprops=wedgeprops,
        radius=0.55,)
ax.legend(wedges[::-1], cans_sort[::-1],
            title='Candidate',
            title_fontsize=20,
            fontsize=20,
            loc='center left',
            bbox_to_anchor=(0.9, 0.5),
            facecolor='lightsteelblue')
fig.tight_layout()
```
