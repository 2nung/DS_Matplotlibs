```python
PI = np.pi
n_point = 300
t = np.linspace(0, 10*PI, n_point)
x = t*np.cos(t)
y = t*np.sin(t)
z = 0.2*t

noise_coeff = 0.1
x_noise = x + noise_coeff*np.random.normal(0, 1, n_point)
y_noise = y + noise_coeff*np.random.normal(0, 1, n_point)
z_noise = z + noise_coeff*np.random.normal(0, 1, n_point)

ax.scatter(x_noise, y_noise, z_noise,s=100)
ax.plot(x, y, z,color='r',linewidth=1)
```

```python
def plot_text(ax, x, y, z, annot, color):
    ax.scatter(x, y, z,s=500,color=color)
    ax.text(x, y, z,annot, color=color, fontsize=30)

plot_text(ax, x_noise[0], y_noise[0], z_noise[0],
            "Start Point", 'b')
plot_text(ax, x_noise[150], y_noise[150], z_noise[150],
            "Middle Point", 'g')
plot_text(ax, x_noise[-1], y_noise[-1], z_noise[-1],
            "End Point", 'r')
```
