```python
x = np.arange(5)
y = np.arange(0, 12, 3)

X, Y = np.meshgrid(x, y)

coordinates = np.empty(X.shape, dtype=object)

for r_idx in range(X.shape[0]):
    for c_idx in range(X.shape[1]):
        x = X[r_idx, c_idx]
        y = Y[r_idx, c_idx]
        coordinates[r_idx, c_idx] = (x, y)

print(coordinates)
```

```python
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')
fig.subplots_adjust(bottom=0, top=1,
                    left=0, right=1)
ax.tick_params(labelsize=20)

norm = plt.Normalize(Z.min(), Z.max())
Z_normed = norm(Z)
facecolors = cm.jet(Z_normed)

surf = ax.plot_surface(X, Y, Z,facecolors=facecolors)
surf.set_facecolor((0,0,0,0))
```

```python
cbar = fig.colorbar(surf,
                    orientation='horizontal',
                    fraction=0.1,
                    pad=0.,
                    shrink=0.9)
cbar_ticks = np.arange(0, 1,0.25)
cbar.set_ticks(cbar_ticks)
cbar.ax.set_xticklabels(cbar_ticks, fontsize=20)
```
