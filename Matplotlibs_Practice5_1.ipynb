```python
import matplotlib.pyplot as plt
import matplotlib.cm as sm
import numpy as np
```

```python
np.random.seed(0)
n_data = 500
data = np.random.normal(0, 1, (n_data,))
fig, ax = plt.subplots(figsize=(14, 10))
ax.tick_params(labelsize=30,
                length=10,
                width=3)
ax.hist(data)
```

```python
np.random.seed(0)
n_data = 500
data = np.random.normal(0, 1, (n_data,))
fig, ax = plt.subplots(figsize=(14, 10))
ax.tick_params(labelsize=30,
               length=10,
                 width=3)
freqs, bin_edges, rects = ax.hist(data)
bin_centers = (bin_edges[:-1] + bin_edges[1:])/2
ax.set_xticks(bin_centers)
ax.set_xticklabels(np.around(bin_centers, 2),
                            rotation=30)
```

```python
ax.hist(data,facecolor='skyblue')
ax.hist(data,facecolor='skyblue',
        edgecolor='darkblue',
        linewidth=3)
ax.hist(data,
        facecolor='skyblue',
        edgecolor='darkblue',
        linewidth=3,
        hatch='//')
```
